```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df= pd.read_excel('https://www.football-data.co.uk/mmz4281/2526/all-euro-data-2025-2026.xlsx')
```

```python
columns_stack=['Div','Date','Time','HomeTeam','AwayTeam','FTHG','FTAG','FTR','HTHG','HTAG','HTR','Referee','HS','AS','HST','AST','HC','AC','HF','AF','HY','AY','HR','AR']
dfn = df[columns_stack]
```

```python
dfn.head()
```

```python
dfn.info()
```

```python
dfn.drop(['Div'], axis=1)
```

```python
sns.countplot(dfn['FTR'])
```

```python
dfn.loc[:, 'FequlsH'] =dfn['FTR'] == dfn['HTR']
```

```python
dfn.FequlsH.value_counts()
```

```python
print(f'The total goals scored till {dfn.Date.iloc[-1].strftime("%Y-%m-%d")} is = {dfn.FTHG.sum() + dfn.FTAG.sum()}')
```

```python
print(f'The total Home goals scored till {dfn.Date.iloc[-1].strftime("%Y-%m-%d")} is = {dfn.FTHG.sum()}')
```

```python
print(f'The total Home goals scored till {dfn.Date.iloc[-1].strftime("%Y-%m-%d")} is = {dfn.FTAG.sum()}')
```
